### Introduction
[Top](#top)

Written Japanese uses a mixed-script system: _kanji_—[morphographic](http://www.ingentaconnect.com/content/jbp/wll/2011/00000014/00000001/art00004) Chinese characters, which is to say: graphemes that mainly represent the morphemes used in word formation—and derived from these, there are two types of phonographic _kana_, which are simpler graphemes representing only the Japanese sound system’s moraic syllable (e.g., ‘_ka_’) foundation, used mostly for grammatical functions. 
<p>
<div align=center><img src="./assets/asahi_codh.jpg">
    <small><i>Figure 1: A text containing archaic handwritten Japanese, <a href="http://www.asahi.com/ajw/articles/AJ201612090010.html">via</a></i></small></div>

Handwritten Japanese before the 19c standardizations of the Meiji era is characterized by myriad, idiosyncratic variations of _kanji_ and _kana_ (_hentaigana_), collectively called _kuzushiji_, and has thereby limited primary access to pre-modern corpora to the few who have attained the necessary literacy (Hashimoto, et al., [2017](http://digitalhumanities.org/dhq/vol/11/1/000281/000281.html)).

Given the value of the historical record, deciphering _kuzushiji_ is a significant [enterprise](https://guides.library.upenn.edu/japanesedigitalresources/kuzushiji), enabled by scanning images of historical texts, with scholars painstakingly hand-labeling characters. A typical practice is training computational models to classify characters, trial-and-error supervised with expert labels acting as the ‘ground truth,’ for corrective feedback. Examples of applications for this topic range from [adapting Edo period recipes](http://www.asahi.com/ajw/articles/AJ201612090010.html) to gathering [seismological](http://www.kyoto-u.ac.jp/cutting-edge/continuum/page01.html) data to better predict earthquakes.

Deep neural networks are state-of-the-art in image recognition, with multiple layers of interconnected ‘neurons’ leveraged to learn numeric representations of training data (e.g., a subset of character images extracted from a corpus, and their manually coded target labels), which can be generalized to evaluate new data (e.g., classifying a novel subset of images and targets). 

For image tasks, the standard architecture is the convolutional neural network (CNN). The key feature of its effectiveness is that a CNN retains the spatial relationships of the pixels comprising images, rather than flattening inputs at the onset. 
<p>
<div align=center>
    <img src="./assets/convnet.jpg" width=625><br>
    <small><i>Figure 2: A typical convolutional neural network, <a href="https://www.mathworks.com/discovery/convolutional-neural-network.html">via</a></i></small></div>

[TensorFlow](https://www.tensorflow.org/) is a high-level library known for providing prefab functions executed over the abstraction of a computational graph. It is an accessible resource which allows diverse researchers to employ deep learning without attending to low-level details (e.g., convolutional operations).

The primary purpose was self-directed learning. This includes the nuanced challenges and rewards of preprocessing multicultural artifacts.

Due to the workload and bugs introduced with this scratch-built, _ad hoc_ approach and a good measure of domain ignorance, this implementation adheres to mostly vanilla structures in order to ease fine-tuning of the hyperparameters and focus on learning important aspects of CNNs.

### Methods
[Top](#top)

Following a standard template, the architecture consists of two convolutional and subsampling layer pairs, followed by a pair of affine layers. Non-linearities for the convolutional layers use an explicitly defined function for leaky rectified linear units (LReLU): the leaky version of this transfer function (intended to prevent the problem of exponentially decreasing gradients) mitigates the issue of cell death in standard ReLU by allowing for a minimal gradient in the negative case, rather than thresholding at zero. The convolutions are downsampled with conventional max-pooling.
<p><div align=center>
    <img src="./assets/tensorboard_graph_zoom.png" width=350><br>
    <small><i>Figure 3: Sample TensorBoard visualization</i></small></div>

Dropout (Srivastava, et al., [2014](https://dl.acm.org/citation.cfm?id=2670313)) is applied to the first affine layer, also from scratch: dropout reduces neuronal co-adaptations by forcing cells to adapt to the more diverse ecologies created by randomly mutated subsets of a network, creating a robust combined model with more independent feature units.

Finally, the readout layer outputs are sent directly to an explicit softmax function with negative log-likelihood loss, to obtain and evaluate a probability distribution for the appropriate class labels. This function is defined such that it scales the logits before normalization, to avoid underflow and overflow, per a best practice referenced by Goodfellow, et al. ([2016](http://www.deeplearningbook.org/contents/numerical.html)) (+[ja](https://www.deeplearningbook.me/)).

Weight decay via L2 regularization is used to help the model better discover underlying patterns in the data with less sensitivity to local noise. Momentum is used with gradient descent, titrating the “velocity” through “friction” rather  than merely “force” (the gradient). Of final note, the training phase employs early-stopping based on the validation set accuracy, aborting when this saturates.

### Load Data

Japanese character data loading and preprocessing.

```python
tf.reset_default_graph()

# =====================================
flist = []
labels = []
layers = ['conv_relu1', 'pool1', 'conv_relu2', 
          'pool2', 'fc1_relu_dropout', 'fc2_readout']
parameters = []
velocities = []
caches = []
rs = 42 # random state
patience = 3
delta = 0.02 # accuracy improvement criterion for early-stopping
impatience = 0
best = -np.Inf
num_epochs = 2
save_model_path = './kuzushijinn/model.ckpt'
ih, iw = 28, 28 # dimensions
c = 1 # channels
trunc = 2000 # number of image samples to limit
char_dir = 'c:/datasets/brsk00000/characters'
# char_dir = 'C:/datasets/200003967/characters'
batch_size = 50
kf = 3 # (height/width), F
fm = 32
fm2 = 64
stride = 1
# =====================================

def get_data(flist, label_dict):
    """Retrieve images and labels.
    
       Use file list to obtain resized grayscale images,
       and corresponding image classes via label dictionary.
       This is a modified function from http://codh.rois.ac.jp.
        
       Args:
           flist: Array of image path strings.
           label_dict: Dictionary:
                       {'char code': (class #, 'char')}
       Returns:
           Numpy arrays x, y containing:
               x: Images [pixel values--intensities here for grayscale], 
               y: Class numbers (integers).
    """
    x = np.zeros((len(flist), ih, iw))
    y = np.zeros((len(flist), 1))
    for idx, file in enumerate(flist):
        img = cv2.imread(file, 0) # as grayscale
        x[idx] = cv2.resize(img, dsize=(ih, iw))
        y_k = os.path.basename(file).split('_')[0] # get char code
        y[idx] = label_dict[y_k][0] # first element of tuple value is char #
    y = np.array(y, dtype='uint8').reshape((len(y), 1))
    return x, y

for root, dirs, files in os.walk(char_dir):
    for directory in dirs:
        labels.append(directory)
    for name in files:
        fpath = os.path.join(root, name)
        flist.append(fpath)
```

We build a dictionary of character code points as keys (strings), with their values being a tuple: containing the class (simply which number in the list of character image directories it is) and the printable representation (ensure you have a Japanese font such as [Noto Sans Japanese](https://www.google.com/get/noto/#sans-jpan) installed--you may need to update the font cache for [Linux](https://thjap.org/linux/ubuntu/5969.html) and [matplotlib](http://bastibe.de/2016-05-30-matplotlib-font-cache.html) if you're newly installing the font).

```python
chars = [chr(int(label[2:], 16)) 
         for label in labels]
label_dict = {label:(idx, chars[idx]) 
              for idx, label in enumerate(labels)}    
```

Next, we split the file list (the images of respective characters extracted from the corpus) into training, validation, and testing lists (60/20/20 by default here, as [recommended](https://www.coursera.org/learn/machine-learning/lecture/QGKbr/model-selection-and-train-validation-sets) by Andrew Ng; this may be far more than necessary to ensure validation data is useful for optimization, to follow [Karpathy](https://github.com/karpathy/char-rnn#best-models-strategy)), and use these to add the image data and class names to their respective arrays.

### Forward Pass
[Top](#top)

Using a memory-cheap size for mini-batches, we'll use tweaked functions provided from TensorFlow's MNIST tutorial to initialize parameters in a way that mitigates saturation. 

I use the rectifier weight initialization to calibrate variances that is described as superior to, say, Xavier, in [this paper](https://arxiv.org/pdf/1502.01852v1.pdf) (He, et al., 2015) and noted in the CS231N [notes](http://cs231n.github.io/neural-networks-2/#init). 

I am also using a small positive constant on the bias to ensure neurons initially fire for ReLU gradient computation, thereby avoiding "dead" units with no gradients: a best practice suggested in the _Deep Learning_ book by Goodfellow, Bengio, and Courville (2016) and implemented in the TensorFlow MNIST tutorial.

```python
def weight_variable(shape):
    with tf.name_scope('weights'):
        initial = tf.truncated_normal(shape, 
                                      stddev = np.sqrt(2.0 / shape[0]), 
                                      name='W')
    return tf.Variable(initial)

def bias_variable(shape):
    with tf.name_scope('biases'):    
        initial = tf.constant(0.1, shape=shape, 
                              name='b')
    return tf.Variable(initial)

def velocity_variable(shape):
    with tf.name_scope('velocities'):        
        initial = tf.zeros(shape, name='v')
    return tf.Variable(initial)
```

We will use dropout for this project. Its utility will depend on the scales we're working at.

```python
with tf.name_scope('input'):
    X = tf.placeholder(tf.float32, [batch_size, ih, iw, c])
    y_ = tf.placeholder(tf.float32, [batch_size, len(labels)])

with tf.name_scope('dropout'):
    keep_prob = tf.placeholder(tf.float32)
    tf.summary.scalar('dropout_keep_prob', keep_prob)    
```

As you may notice in the functions below, I will attempt to ease convolutional operations' expense with the im2col/col2im trick.

<div align=center><img src="./assets/im2col.png" height=252 width=550>
<img src="./assets/im2colmult_.png" height=252 width=550>
<small><em>im2col & matmul(W_row, X_col) operation (result = num_kernels × num_patches), via <a href="https://petewarden.com/2015/04/20/why-gemm-is-at-the-heart-of-deep-learning/">Why GEMM is at the heart of deep learning</a></em></small></div>

```python
def variable_summaries(var):
    # Via TensorFlow tutorial:
    # https://www.tensorflow.org/tutorials/mnist/pros/#deep-mnist-for-experts    
    with tf.name_scope('summaries'):
        mean = tf.reduce_mean(var)
        tf.summary.scalar('mean', mean)
    with tf.name_scope('stddev'):
        stddev = tf.sqrt(tf.reduce_mean(tf.square(var - mean)))
    tf.summary.scalar('stddev', stddev)
    tf.summary.scalar('max', tf.reduce_max(var))
    tf.summary.scalar('min', tf.reduce_min(var))
    tf.summary.histogram('histogram', var)    
    
def get_dims(h, h2, w, w2, s, p):
    # (W - F + 2P/S + 1)
    out_h = (h - h2 + 2*p) // s + 1
    out_w = (w - w2 + 2*p) // s + 1    
    return out_h, out_w    
    
def im2col(X, h, w, d, out_h, out_w, s=1, p=1):
    # TensorFlow conversion of CS231N Numpy functions in im2col.py: 
    # http://vision.stanford.edu/teaching/cs231n/assignments.html
    # http://cs231n.github.io/assignments2017/assignment2/
    # http://cs231n.stanford.edu/assignments/2017/spring1617_assignment2.zip    
    X_p = tf.pad(X, [[0, 0], [p, p], [p, p], [0, 0]])
    # Slice offsets: s*oh:s*oh+kh, s*ow:s*ow+kw
    # Accounting for stride, moving from along output dims, excise
    # kernel-sized tiles from the input.    
    X_l = [tf.slice(X_p, [0, s*oh, s*ow, 0], [-1, h, w, -1])
           for oh in range(out_h) for ow in range(out_w)]
    # In a list of length: flattened output dims (# locations w/ s=1, p=0), 
    # we now have a tensor of kernel dims w/ 
    # no. samples, e.g., (100, 3, 3, 1)
    # Make into tensor of dims: kernel height, kernel width, kernel depth, 
    # (no. samples--e.g., batch_size), (output dims--e.g., 28*28=784)
    # e.g., 3 × 3 × 1 × 100 × 784
    X_mt = tf.transpose(tf.stack(X_l, axis=1), [2, 3, 4, 0, 1])
    # Shape into matrix:
    # Each column vector magnitude = flattened kernel dims (e.g., 3*3*1)
    # Each row vector magnitude = # of locations for the images (e.g., 100*784)
    # So each coordinate = image reflection in a kernel dimension   
    X_col = tf.reshape(X_mt, [d*h*w, -1])
    # Now we have a, for example, (3*3*1) × (100*28x*28) tensor to be transformed
    # k_dims_flat × num_locations
    return X_col    
    
def convolve_forward(X, W, b, stride=1, padding=1):
    # convolve == matrix multiply
    n, h, w, _ = X.get_shape().as_list()    
    kh, kw, kd, km = W.get_shape().as_list()
    # (W - F + 2P/S + 1)
    out_h, out_w = get_dims(h, kh, w, kw, 
                            stride, padding)
    X_col = im2col(X, kh, kw, kd, 
                   out_h, out_w, 
                   stride, padding)
    # transpose, shape weights into matrix (e.g., 32 × 3*3*1)
    W_row = tf.reshape(tf.transpose(W, [3, 0, 1, 2]), 
                       [km, kh*kw*kd])
    # (32 × 9) * (9 × 78400)
    z = tf.matmul(W_row, X_col) + tf.reshape(b, [-1, 1])
    # reshape to proper dimensions--e.g., 32 × 100 × 28 × 28
    z = tf.reshape(z, [km, n, out_h, out_w] )
    # transpose to n, out_h, out_w, km
    # 100 × 28 × 28 × 32
    z = tf.transpose(z, [1, 2, 3, 0])
    cache = (X, X_col, W, W_row, stride, padding)
    return z, cache

def leaky_relu(X, eps=0.2):
    return tf.maximum(X, eps * X), (X, eps)

def max_pool_forward(X, pool_h=2, pool_w=2, stride=2, padding=0):
    n, h, w, km = X.get_shape().as_list()
    # 100 × 28 × 28 × 32
    out_h, out_w = get_dims(h, pool_h, w, pool_w, stride, padding)
    X_col = im2col(X, pool_h, pool_w, km, 
                     out_h, out_w, stride, padding)
    # X_col example shape: (2*2*32) × (100*14*14)
    prop_d = [pool_h * pool_w, km, n, out_h, out_w]
    X_col = tf.reshape(X_col, prop_d)
    # X_col now would be [4 × 32 × 100 × 14 × 14]
    pool = tf.reduce_max(X_col, axis=0)
    # pool: [32 × 100 × 14 × 14]
    # transpose pool to: [100 × 14 × 14 × 32] or: [n, h, w, m]
    pool = tf.transpose(pool, [1, 2, 3, 0])
    return (pool, 
            (X, X_col, pool, pool_h, pool_w), n)

def affine_forward(X, W, b, n):
    X_fc = tf.reshape(X, [n, -1])
    z = tf.matmul(X_fc, W) + b
    return z, (X_fc, W)

def dropout_forward(X, keep_prob=0.5):
    # Attempting to implement inverted dropout
    # The binary mask selects a percentage of values to be masked to 0
    # (e.g., keep_prob > 50% means random floats added to it 
    # can be smaller and still likely end up > 1,
    # rather than being 50% likely or more to be closer to 0); 
    # if keep_prob isn't 0 (avoid zero division),
    # we scale up the remaining training phase activations by keep_prob 
    # so that the val/test phase activations without
    # dropout don't have to be scaled down to account for 
    # their undropped neuron outputs.
    mask = tf.floor(tf.random_uniform(tf.shape(X), dtype=X.dtype) + keep_prob)
    mask *= (1. / keep_prob if keep_prob is not None else 0.)
    return X * mask, mask

def softmax_cross_entropy(logits, y_):
    # numerically stable
    logits_scaled_exp = tf.exp(logits - tf.reduce_max(logits))
    y = tf.transpose(tf.transpose(logits_scaled_exp) 
                            / tf.reduce_sum(logits_scaled_exp, axis=1))
    with tf.name_scope('total'):
            cross_entropy = -tf.reduce_mean(tf.reduce_sum(y_ * tf.log(y), 
                                                                 axis=1))
    return y, cross_entropy
    
def forward_pass(X, layers, parameters, velocities, s=1, p=1):
    caches = dict()
    for layer in layers:
        with tf.name_scope(layer + '_forward'):
            if layer == 'conv_relu1':
                with tf.variable_scope('conv_relu1') as scope:
                    W1 = weight_variable([kf, kf, c, fm])
                    vW1 = velocity_variable([kf, kf, c, fm])    
                    b1 = bias_variable([fm]) 
                    vb1 = velocity_variable([fm])
                
                variable_summaries(W1)
                variable_summaries(b1)
                
                parameters.append(W1)
                parameters.append(b1)
                velocities.append(vW1)
                velocities.append(vb1)
                
                with tf.name_scope('conv1_preactivations'):
                    conv1_layer, conv1_cache = convolve_forward(X, W1, b1, 
                                                              stride=s, 
                                                              padding=p)
                    tf.summary.histogram('conv1_preactivations', conv1_layer)
                with tf.name_scope('conv1_activations'):
                    conv1_activation_layer, conv1_relu_cache = leaky_relu(conv1_layer)
                    tf.summary.histogram('conv1_activations', 
                                         conv1_activation_layer)
                
                caches['conv_relu1'] = (conv1_cache, conv1_relu_cache)
                
            if layer == 'pool1':
                pool1, pool1_cache, n = max_pool_forward(conv1_activation_layer)
                _, pool1_h, pool1_w, pm1 = pool1.get_shape().as_list()
                
                caches['pool1'] = pool1_cache
                
            if layer == 'conv_relu2':
                with tf.variable_scope('conv_relu2') as scope:
                    W2 = weight_variable([kf, kf, pm1, fm2])
                    vW2 = velocity_variable([kf, kf, pm1, fm2])
                    b2 = bias_variable([fm2])
                    vb2 = velocity_variable([fm2])  

                variable_summaries(W2)
                variable_summaries(b2)

                parameters.append(W2)
                parameters.append(b2)
                velocities.append(vW2)
                velocities.append(vb2)
                
                with tf.name_scope('conv2_preactivations'):
                    conv2_layer, conv2_cache = convolve_forward(pool1, W2, b2, 
                                                              stride=s, 
                                                              padding=p)
                    tf.summary.histogram('conv2_preactivations', conv2_layer)
                with tf.name_scope('conv2_activations'):
                    conv2_activation_layer, conv2_relu_cache = leaky_relu(conv2_layer)
                    tf.summary.histogram('conv2_activations', 
                                         conv2_activation_layer)
                
                caches['conv_relu2'] = (conv2_cache, conv2_relu_cache)
                
            if layer == 'pool2':
                pool2, pool2_cache, n = max_pool_forward(conv2_activation_layer)
                _, pool2_h, pool2_w, pm2 = pool2.get_shape().as_list()
                
                hidden_nodes = 1024
                flat = [pool2_h*pool2_w*pm2, hidden_nodes]
                with tf.variable_scope('pool2') as scope:                
                    W3 = weight_variable(flat)
                    vW3 = velocity_variable(flat)
                    b3 = bias_variable([hidden_nodes])
                    vb3 = velocity_variable([hidden_nodes])  

                variable_summaries(W3)
                variable_summaries(b3)

                parameters.append(W3)
                parameters.append(b3)
                velocities.append(vW3)
                velocities.append(vb3)   
                
                caches['pool2'] = pool2_cache                  
                
            if layer == 'fc1_relu_dropout':
                with tf.name_scope('fc_preactivations'):
                    fc1, fc1_cache = affine_forward(pool2, W3, b3, n)                
                    tf.summary.histogram('fc_pre_activations', fc1)
                with tf.name_scope('fc_activations'):                    
                    fc1_activation, fc1_relu_cache = leaky_relu(fc1)  
                    tf.summary.histogram('fc_activations', fc1_activation)    
                with tf.name_scope('fc_dropout'):                
                    fc1_dropped, fc1_dropout_cache = dropout_forward(fc1_activation, 
                                                                keep_prob) 
                    tf.summary.histogram('fc_dropout', fc1_dropped) 
                output_nodes = len(labels) 
                
                with tf.variable_scope('fc1_relu_dropout') as scope:                
                    W4 = weight_variable([hidden_nodes, output_nodes])
                    vW4 = velocity_variable([hidden_nodes, 
                                        output_nodes])    
                    b4 = bias_variable([output_nodes])
                    vb4 = velocity_variable([output_nodes])    
                
                variable_summaries(W4)
                variable_summaries(b4)  
                
                parameters.append(W4)
                parameters.append(b4)
                velocities.append(vW4)
                velocities.append(vb4)   
            
                caches['fc1_relu_dropout'] = (fc1_cache, fc1_relu_cache, 
                                              fc1_dropout_cache)
                
            if layer == 'fc2_readout':
                
                with tf.name_scope('fc2'):
                    fc2, fc2_cache = affine_forward(fc1_dropped, W4, b4, n)                
                caches['fc2_readout'] = fc2_cache
                tf.summary.histogram('fc2', fc2)   
                    
                with tf.name_scope('cross_entropy'):                    
                    y, loss = softmax_cross_entropy(fc2, y_)
                    tf.summary.scalar('cross_entropy', loss)             
            
    return parameters, velocities, y, loss, caches

parameters, velocities, y, loss, caches = forward_pass(X, layers, 
                                                       parameters, velocities, 
                                                       stride, padding)     
```

The use of caches for backpropagation is a commonplace, recommended in the CS231N optimization [notes](http://cs231n.github.io/optimization-2/#backprop-in-practice-staged-computation).

### Backward Pass

The col2im and backwards convolution and pooling operations need work, they are particular to this specific project, and likely flawed, although somehow the network still learns. Likely in need of [gradient checking](http://ufldl.stanford.edu/tutorial/supervised/DebuggingGradientChecking/) and needs debugging of histogram NaN errors found when running this network with larger sets on AWS/GPU.

Due to missing features in TensorFlow as compared with NumPy (regarding indexed assignment), I invented a cantrip of sorts to achieve col2im-like goals in the backward pooling function.

```python
def col2im(delta, kh, kw, kd, n, oh, ow, stride, padding):
    oh = oh + (2 * padding)
    ow = ow + (2 * padding)
    X_im = im2col(delta, kh, kw, kd, oh, ow, stride, 2 * padding)
    X_im = tf.reshape(X_im, [kh*kw*kd, -1, oh, ow])
    X_im = tf.transpose(X_im, [1, 2, 3, 0])    
    if padding > 0:
        X_im = X_im[:, padding:oh-padding, padding:ow-padding, :] 
    return X_im
    
def affine_backward(delta, cache):
    X, W = cache
    h_delta_tmp = tf.matmul(delta, tf.transpose(W))
    ndW = tf.matmul(tf.transpose(X), delta)
    ndb = tf.reduce_sum(delta, axis=0)
    return h_delta_tmp, ndW, ndb

def leaky_relu_prime(cache):
    # Values are ϵ if x is negative, 1 otherwise.
    X, eps = cache
    dX = tf.where(X < 0, 
                  tf.constant(eps, shape=X.shape), 
                  tf.ones(shape=X.shape))
    return dX

def pool_backward(delta, cache):
    X, X_col, pool, pool_h, pool_w = cache
    n, h, w, c = X.get_shape().as_list()
    pn, ph, pw, pm = pool.get_shape().as_list()
    delta_r = tf.reshape(delta, [n, ph, pw, pm])
    dX_col = tf.zeros_like(X_col)
    dX_col = tf.transpose(dX_col, [0, 2, 3, 4, 1]) 
    X_col2 = tf.transpose(X_col, [0, 2, 3, 4, 1])
    pool_filter = X_col2 + pool
    pool_filter = pool_filter / 2
    pool_cond = tf.equal(tf.cast(X_col2, dtype=pool_filter.dtype), 
                         pool_filter)
    dX_im = tf.where(pool_cond, tf.ones_like(X_col2), dX_col)
    dX_im = delta_r * dX_im
    dX_im = tf.transpose(dX_im, [1, 2, 3, 4, 0])
    dX_im = tf.reshape(dX_im, [n, pool_h*ph, pool_w*pw, c])
    # dX_im should now be expanded tensor of zeros,
    # blocking the error signal, except where
    # error indices match max pool indices;
    # so delta has been expanded to proper dims. 
    return dX_im

def conv_backward(delta, cache):
    X, X_col, W, W_row, stride, padding = cache
    n, h, w, c = X.get_shape().as_list()
    kh, kw, kd, km = W.get_shape().as_list()
    dX_im = col2im(delta, kh, kw, kd, n, h, w, stride, padding)
    dX_im = tf.reshape(dX_im, [km, -1])
    dX_im = tf.matmul(tf.transpose(dX_im), W_row)
    dX_im  = tf.reshape(dX_im, [n, h, w, -1, c])
    dX_im = tf.reduce_max(dX_im, axis=3)
    dX_col = tf.reshape(delta, [-1, km])
    ndW = tf.matmul(X_col, dX_col)
    ndW = tf.reshape(ndW, [kh, kw, kd, km])
    ndb = tf.reduce_sum(delta, axis=(0, 1, 2))
    return dX_im, ndW, ndb

def backward_pass(y, layers, caches):
    gradients = []
    with tf.name_scope('out_delta'):             
        error = (y - y_) / batch_size        
        tf.summary.histogram('out_delta', error)     

    for layer in layers[::-1]:
        with tf.name_scope(layer + '_back'):
            if layer == 'fc2_readout':
                with tf.name_scope('fc2_delta'):
                    h_delta_tmp, ndW4, ndb4 = affine_backward(error, 
                                                              (caches['fc1_relu_dropout'][1][0], 
                                                               caches['fc2_readout'][1]))
                    h_delta4 = tf.multiply(h_delta_tmp, 
                                          leaky_relu_prime(caches['fc1_relu_dropout'][1]))    
                    # Dropout adjustment
                    h_delta4 *= caches['fc1_relu_dropout'][2]
                    tf.summary.histogram('fc2_delta', h_delta4)

                    variable_summaries(ndb4)
                    variable_summaries(ndW4)

                    gradients.append(ndb4)
                    gradients.append(ndW4)     

            if layer == 'fc1_relu_dropout':
                with tf.name_scope('fc1_delta'):
                    h_delta3, ndW3, ndb3 = affine_backward(h_delta4, caches['fc1_relu_dropout'][0])    
                    tf.summary.histogram('fc1_delta', h_delta3)

                    variable_summaries(ndb3)
                    variable_summaries(ndW3)

                    gradients.append(ndb3)
                    gradients.append(ndW3)  
                    
            if layer == 'pool2':
                with tf.name_scope('pool2_delta'):                
                    h_delta_tmp = pool_backward(h_delta3, caches['pool2'])
                    h_delta2 = tf.multiply(h_delta_tmp, 
                                          leaky_relu_prime(caches['conv_relu2'][1]))              
                    tf.summary.histogram('pool2_delta', h_delta2)  
                    
            if layer == 'conv_relu2':
                with tf.name_scope('conv2_delta'):                   
                    dX_im2, ndW2, ndb2 = conv_backward(h_delta2, caches['conv_relu2'][0])
                    tf.summary.histogram('conv2_delta', dX_im2)       
                    
                    variable_summaries(ndb2)
                    variable_summaries(ndW2)

                    gradients.append(ndb2)
                    gradients.append(ndW2)                       
                
            if layer == 'pool1':
                with tf.name_scope('pool1_delta'):                
                    h_delta_tmp = pool_backward(dX_im2, caches['pool1'])
                    h_delta1 = tf.multiply(h_delta_tmp, 
                                          leaky_relu_prime(caches['conv_relu1'][1]))              
                    tf.summary.histogram('pool1_delta', h_delta1)   
                    
            if layer == 'conv_relu1':
                with tf.name_scope('conv1_delta'):                   
                    dX_im, ndW1, ndb1 = conv_backward(h_delta1, caches['conv_relu1'][0])
                    tf.summary.histogram('conv1_delta', dX_im)       
                    
                    variable_summaries(ndb1)
                    variable_summaries(ndW1)

                    gradients.append(ndb1)
                    gradients.append(ndW1)   
                    
    return gradients

gradients = backward_pass(y, layers, caches)
```

### Run Network
[Top](#top)

```python
# L2 regularization
loss += 0.5 * (lmbda / X_train.shape[0]) * tf.reduce_sum([tf.norm(tf.square(parameter)) 
                                                 for parameter in parameters])

# Stochastic gradient descent over mini-batches w/ vanilla momentum, L2
def sgd_with_momentum(velocities, parameters, gradients, eta=0.001, mu=0.9):
    return [parameter.assign((1 - eta * (lmbda / X_train.shape[0])) *
                             parameter + v.assign(mu * v - 
                                                  ((eta / batch_size) * gradient))) 
            for parameter, gradient, v in zip(parameters, 
                                              gradients[::-1], 
                                              velocities)]
      
with tf.name_scope('train'):
    trainer = tf.group(*sgd_with_momentum(velocities, parameters, gradients, 
                                          eta=0.01, mu=0.5))   
    
with tf.name_scope('accuracy'):
    pred = tf.argmax(y, axis=1)
    act = tf.argmax(y_, axis=1)
    with tf.name_scope('correct_prediction'):
        correct_prediction = tf.cast(tf.equal(pred, act), tf.float32)
    with tf.name_scope('accuracy'):
        accuracy = tf.reduce_mean(correct_prediction)    
    tf.summary.scalar('accuracy', accuracy)           
```

The early-stopping is based on the validation accuracy; the training graph and split accuracies are available via TensorBoard.

```python
def get_n_batches(X):
    return int(X.shape[0] // batch_size)

def get_mini_batches(X, y_):
    num_batches = get_n_batches(X)
    for i in range(0, num_batches * batch_size, batch_size):
        if i == (batch_size * (num_batches - 1)):
            # last batch, fill w/ samples left
            if len(X[i:]) <= batch_size:
                xt, yt = X[i:], y_[i:]
        else:
            xt, yt = X[i:batch_size+i], y_[i:batch_size+i]
    yield xt, yt   

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    # Merge all summaries and write to ./tmp
    # terminal: tensorboard --logdir=:tmp
    # browser: localhost:6006
    
    merged = tf.summary.merge_all()
    train_writer = tf.summary.FileWriter('./tmp' + '/train', sess.graph)
    val_writer = tf.summary.FileWriter('./tmp' + '/val')  
    test_writer = tf.summary.FileWriter('./tmp' + '/test')  

    for epoch in range(num_epochs):
        avg_cost = 0
        train_X, train_y = shuffle(X_train, y_train, 
                                   random_state=rs)
        test_X, test_y = shuffle(X_test, y_test, 
                                   random_state=rs)
        val_X, val_y = shuffle(X_val, y_val, 
                                   random_state=rs)
        # Training
        for k in tqdm_notebook(range(get_n_batches(train_X))):
            start = k * batch_size
            end = start + batch_size
            xt = train_X[start:end]
            yt = train_y[start:end]
            summary, _ = sess.run([merged, trainer], 
                                  feed_dict={X:xt, y_:yt, keep_prob: 0.5})
            cost = loss.eval(feed_dict={X:xt, y_:yt, keep_prob: 1.0})
            avg_cost += cost / get_n_batches(train_X)
            
            train_writer.add_summary(summary, _)
            train_writer.flush()
        
        # Evaluate validation set accuracy
        avg_val_acc = 0   
        avg_test_acc = 0
        
        for val_x, val_t in get_mini_batches(val_X, val_y):        
            summary, val_acc = sess.run([merged, accuracy], 
                                        feed_dict={X:val_x, y_:val_t, 
                                                   keep_prob: 1.0})
            avg_val_acc += val_acc
            
            val_writer.add_summary(summary, val_acc)
            val_writer.flush()
        
        avg_val_acc = np.mean(avg_val_acc)

        # Evaluate test set accuracy            
        for test_x, test_t in get_mini_batches(test_X, test_y):
            summary, test_acc = sess.run([merged, accuracy], 
                                         feed_dict={X:test_x, y_:test_t, 
                                                    keep_prob: 1.0})                
            avg_test_acc += test_acc
            
            test_writer.add_summary(summary, test_acc)
            test_writer.flush()
            
        avg_test_acc = np.mean(avg_test_acc)

        # Print statistics
        print("Epoch: {:d}".format(epoch + 1), 
              "Training cost: {:.3f}".format(avg_cost), 
              "Validation Accuracy = {:.2f}%".format(avg_val_acc),
              "Test Accuracy = {:.2f}%".format(avg_test_acc), 
              flush=True)          
                 
        # Early-stopping
        if epoch > 0:
            if np.round(avg_val_acc - delta, 2) > best:
                best = np.round(avg_val_acc, 2)
                impatience = 0
            else:
                impatience += 1
        if impatience >= patience or np.round(avg_val_acc, 2) >= 0.97:
            print("Early stopping at epoch {:d}".format(epoch + 1), 
                  "with validation accuracy at {:.2f}%".format(avg_val_acc),
                  "and patience {:d}.".format(patience))
            break        
            
    # Save model
    saver = tf.train.Saver()
    save_path = saver.save(sess, save_model_path)
    print("Model saved in file: %s" % save_path)          
```

In the future, I would like to experiment in a similar fashion with the modern [ETL Character Database](http://etlcdb.db.aist.go.jp/), using Geoffrey Hinton’s [capsule](https://www.youtube.com/watch?v=pPN8d0E3900) architecture—which I had encountered through one of his recorded talks prior to CNNs when I first began my studies in this area—now that new papers have been published to detail the work of researchers such as [Sara Sabour](https://www.nytimes.com/2017/11/28/technology/artificial-intelligence-research-toronto.html), and various [explainers](https://medium.com/@pechyonkin/understanding-hintons-capsule-networks-part-i-intuition-b4b559d1159b) and code [repositories](https://github.com/naturomics/CapsNet-Tensorflow) have been released.
    
My own mental models of reading Japanese stem from research I had previously read from authors such as [Koji Miwa](http://kojimiwa.com/publication.html) and [Harald Baayen](http://www.sfs.uni-tuebingen.de/~hbaayen/index.html) (known for his grammatical work, as well). To fully explore the applications of such biological models to network design, integrating “[attention](http://bair.berkeley.edu/blog/2017/11/09/learn-to-attend-fovea/)” and some notion of priority in layer-to-layer activations might be necessary, such that one might quantify the character-driven processing and submorphemic supplementation thereof.

There have been putatively physiological “[elastic distortions](http://ieeexplore.ieee.org/document/1227801/?part=1%7Csec2#sec2)” used to artificially expand MNIST data; I have considered a more explicit modeling of such a process, as it seems a promising avenue, given the sensorimotor complexities of _kanji_ handwriting; recent RNN research on [calligraphic stylisation](https://vimeo.com/244227371) could be helpful for this.

For this project there are numerous improvements to be made. A basic function for feature visualization by looking at the convolutional layers' activations.

```python
def display_activations(layer_acts):
    filters = len(layer_acts[0][0][0])
    lid = len(layer_acts[0]) #ih, iw
    rngc = len(layer_acts) # In case we want them all.
    rows = np.sqrt(filters)    # Presumably, filters divisible by 8...
    fig = plt.figure(figsize=(16, 16))
    fig.suptitle('Layer activations') 
    for i in range(rngc):      
        # For each image...
        for j in range(filters):
            # For each filter...
            plt.subplot(rows, np.ceil(filters / rows) + 1, j+1)   
            # Add to grid:
            # All pixel values/intensities for the image's current filter...
            # Reshaped and converted to image.
            plt.imshow(np.reshape(layer_acts[i][:, :, j], [lid, lid]), 
                       interpolation="nearest", cmap="gray")
            # plt.title('Img: ' + str(i + 1) + '\n Filt: ' + str(j + 1))    
            plt.title('Image: ' + chars[acts[i]] + '\n Filter: ' + str(j + 1))
            plt.axis("off")
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)     
        # Just save first, for now...
        acts_img_path = './layer_acts_{:d}.png'.format(i+1)
        if i == 0 and not os.path.exists(acts_img_path):
                plt.savefig(acts_img_path, dpi=150, transparent=False)             
        plt.show()   
```

The results visualization is a mere sample plot for a single batch at present.

```python
def display_results(chars, preds, acts):   
    # Plot results
    with plt.style.context("seaborn-talk"):    
        plt.rcParams['font.family'] = 'Noto Sans CJK JP Regular'
        fig = plt.figure()
        fig.suptitle('Actual vs. Expected')
        for rng, idx in zip(range(20), range(20)):
            plt.subplot(4, 5, rng + 1)
            img = test_x[idx]
            plt.imshow(np.reshape(img, [ih, iw]), 
                       interpolation="gaussian", cmap="gray")
            plt.title("y_: %s, y: %s" % (chars[acts[idx]], 
                                        chars[preds[idx]]))
            plt.axis("off")
        plt.tight_layout()
        plt.subplots_adjust(top=0.9)    
        res_img_path = './moji_res_{:s}.png'.format(labels[acts[idx]])
        if i == 0 and not os.path.exists(res_img_path):    
            plt.savefig(res_img_path, dpi=150, transparent=False)
        plt.show()

display_results(chars, preds, acts)
```

An attempt at examining the softmax results for the top _k_ possibilities for a given image.

```python
correct = True # Manually toggle which version to look at...
for img_ix in range(batch_size):
    if not correct:
        if labels[acts[img_ix]] != labels[preds[img_ix]]:
            # Select incorrect predictions...
            break
    else:
        if labels[acts[img_ix]] == labels[preds[img_ix]]:
            # Select correct predictions...
            break   
k = 3
top_k = [(idx, val) for idx, val in enumerate(sma[0][img_ix])
        if val in heapq.nlargest(k, sma[0][img_ix])]

def get_softmax(top_k, img_ix):         
    with plt.style.context("seaborn-talk"):
        plt.rcParams['font.family'] = 'Noto Sans Japanese' 
        fig, axes = plt.subplots(1, 2)
        plt.suptitle('Top {:d} Softmax Candidates'.format(len(top_k)))
        img = test_x[img_ix]
        axes[0].imshow(np.reshape(img, [ih, iw]), 
                       interpolation="gaussian", cmap="gray")
        axes[0].set_title("Actual: %s" % (chars[acts[img_ix]]))
        axes[0].set_axis_off()
        axes[1].set_title("Predicted: %s" % (chars[preds[img_ix]]))
        axes[1].barh(np.arange(k), 
                     [j for i, j in top_k])
        plt.yticks(np.arange(k), 
                   [chars[i] 
                    for i, j in top_k]) # labels[i] for code points
        fig.tight_layout()
        plt.subplots_adjust(top=0.9) 
        soft_img_path = './soft_res_{:s}.png'.format(labels[acts[img_ix]])
        if i == 0 and not os.path.exists(soft_img_path):    
            plt.savefig(soft_img_path, dpi=150, transparent=False)
        plt.show()

get_softmax(top_k, img_ix)
```
